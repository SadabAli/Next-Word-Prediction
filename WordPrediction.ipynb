```python
faqs='''Islam is a monotheistic Abrahamic religion founded on the teachings of the Prophet Muhammad, who is considered the last prophet in a long line of prophets, including Adam, Noah, Abraham, Moses, and Jesus. Islam's central scripture, the Qur'an, is believed by Muslims to be the literal word of God (Allah in Arabic) as revealed to Muhammad through the angel Gabriel. With over 1.9 billion followers worldwide, Islam is the second-largest religion in the world after Christianity.
Core Beliefs of Islam
Islam is based on a set of core beliefs and practices that define the faith and the lifestyle of Muslims:

Monotheism: At the heart of Islam is the belief in a single, omnipotent, omniscient, and merciful God. This concept of monotheism is known as Tawhid. Allah is considered to be the creator and sustainer of the universe, and nothing is comparable to Him.

Prophethood: Muslims believe that God has sent a series of prophets to guide humanity throughout history. The prophets are considered messengers who brought God's word to different peoples. Muhammad is regarded as the "Seal of the Prophets," meaning he is the final prophet, completing the message that began with previous prophets.

The Qur'an: The Qur'an is the holy book of Islam, believed to be the final and unaltered word of God. It is written in Arabic and divided into 114 chapters (called Surahs). The Qur'an covers various aspects of life, including theology, law, morality, guidance, and personal conduct.

Angels: Muslims believe in the existence of angels, who are considered messengers and servants of God. Angels are believed to be created from light and have various roles, such as recording human deeds, delivering messages, and performing other duties assigned by God.

The Afterlife: Islam teaches that life on Earth is a test and that there is an afterlife where individuals will be judged by God based on their deeds and intentions. The righteous will be rewarded with paradise (Jannah), while those who led a life of sin and disbelief will be punished in hell (Jahannam).

Divine Predestination: Islam teaches that God has knowledge of all things and that everything happens according to His divine will and plan. This concept is known as Qadar.'''
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index
```

```python
for sectence in faqs.split('\n'):
    print(sectence)
```

```python
for sectence in faqs.split('\n'):
    print(tokenizer.texts_to_sequences([sectence])[0])
```

```python
input_sequences = []
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```
